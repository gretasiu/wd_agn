## Bondi accretion of WD in AGN

```python
# --- Bondi accretion function ---
def bondi_accretion_rate(M, rho, lam, G, v, c_s):
    v_eff = (v**2 + c_s**2)**(3/2)
    #v_eff = v**3
    mdot = (lam * 4 * np.pi * G**2 * M**2 * rho) / v_eff  # [g/s]
    return mdot

# --- Blandford-Znajek jet luminosity ---
def L_BZjet(eta_bar, mdot):
    #eta_j = a_BH**2
    L_jet = eta_bar*mdot*(c**2) # Tagawa 2023 eqn 1
    return L_jet

# --- Accretion power ---
def L_acc(eta, mdot):
    L = eta*mdot*(c**2) 
    return L

def stromgren_radius(L_ion, n_H, alpha_B=2.6e-13, hv_avg=(13.6) * 1.60218e-12):
    Q = L_ion / hv_avg  # Ionizing photon rate [s^-1]
    n = n_H             # [cm^-3]
    R_s_cm = ((3 * Q) / (4 * np.pi * n**2 * alpha_B))**(1 / 3)
    R_s_pc = R_s_cm / 3.086e18  # Convert to parsecs
    return R_s_pc
```
